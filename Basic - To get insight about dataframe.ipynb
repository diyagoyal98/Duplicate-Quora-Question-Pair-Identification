```python
#finding count of duplicate questions in the whole dataset
qid=pd.Series(df['qid1'].tolist()+df['qid2'].tolist())
#print(qid)
print("Number of unique questons  ",np.unique(qid).shape[0])

x=qid.value_counts()>1
print("Number of duplicates ",x[x].shape[0])
```
